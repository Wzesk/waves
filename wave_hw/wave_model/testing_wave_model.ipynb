# Process
1. Import dependencies including the wave model
2. create a wave (my_wave) object and print the calculated wavelength as a test
3. Create arrays and simulate velocities at different depths
4. Visualize the relationship between velocities and depth to confirm they diminish as expected as depth approachs the bottom

```python
import numpy as np
import wave_model as wm
import matplotlib.pyplot as plt
```

```python
my_wave = wm.wave_solver(10,75,1) #starting with some test values -- period of 10, amplitude of 1
my_wave.initial_wavelength
```

```python
wavelength = float(my_wave.initial_wavelength)
wavenumber = float(my_wave.initial_wavenumber)
wavefrequency = float(my_wave.wavefrequency)

horizvelocities = [0,0,0,0] 
vertvelocities = [0,0,0,0] 
pressure = [0,0,0,0] 

horizvelocities[0] = float(my_wave.horizontal_velocity(-1))
vertvelocities[0] = float(my_wave.vertical_velocity(-1))
pressure[0] = float(my_wave.wave_pressure(-1))

horizvelocities[1] = float(my_wave.horizontal_velocity(-25))
vertvelocities[1] = float(my_wave.vertical_velocity(-25))
pressure[1] = float(my_wave.wave_pressure(-25))

horizvelocities[2] = float(my_wave.horizontal_velocity(-50))
vertvelocities[2] = float(my_wave.vertical_velocity(-50))
pressure[2] = float(my_wave.wave_pressure(-50))

horizvelocities[3] = float(my_wave.horizontal_velocity(-74))
vertvelocities[3] = float(my_wave.vertical_velocity(-74))
pressure[3] = float(my_wave.wave_pressure(-74))
```

```python
plt.title("Velocity Diminishing with Depth (testing wave model)")
plt.xlabel("Measurement Depth")
plt.ylabel("Max Acceleration")
plt.plot([1,25,50,74],horizvelocities,color="blue",label="Horizontal")
plt.plot([1,25,50,74],vertvelocities,color="green",label="Vertical")
plt.show()
```

```python
fig, ((axs1)) = plt.subplots(1, 1,figsize=(12, 6))
fig.tight_layout(h_pad=5,w_pad=5)

axs1.set_title("Max Horizontal and Vertical Velocity")

colors = ["blue","green","red"]

axs1.plot([1,25,50,74],horizvelocities,color=colors[0],label="Horizontal Velocity")
axs1.plot([1,25,50,74],vertvelocities,color=colors[1],label="Vertical Velocity")

for axs in fig.get_axes():
    axs.set(xlabel="Water Depth (meters)", ylabel="")
    #axs.set_ylim(0, 1)
    #axs.set_xlim(1, 80)
    axs.autoscale()
    axs.legend(loc='upper right')
    
plt.show()
```
